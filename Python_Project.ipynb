```python
# Seprating out the categorical and continuous variables
def seprate_data_types(df):
    categorical = []
    continuous = []
    for column in df.columns:                # looping on the number of columns
        if df[column].dtype== object:     
                                                   
            categorical.append(column)
        else:
            continuous.append(column)
            
    return categorical, continuous 


categorical, continuous = seprate_data_types(df)         # Calling the function


table = [categorical, continuous]
print(tabulate({"Categorical":categorical,
                "continuous": continuous}, headers = ["categorical", "continuous"]))
```

Task 8 Identify and display the names of all production companies along with the number of times they appear in the dataset.

```python
import json
# Convert the string representation of production companies into a list of dictionaries
df['production_companies'] = df['production_companies'].apply(lambda x: json.loads(x))
```

```python
# # Initialize an empty dictionary to store production company names and their counts
production_counts = {}

# Iterate over each row in the DataFrame
for row in df.itertuples(index=False):
    # Iterate over each production company in the row
    for company in row.production_companies:
        # Extract the production company name
        company_name = company['name']
        
        # Increment the count for the production company
        if company_name in production_counts:
            production_counts[company_name] += 1
        else:
            production_counts[company_name] = 1

# Create a new DataFrame to store the production company names and their counts
production_df = pd.DataFrame({'Production Company': list(production_counts.keys()), 'Count': list(production_counts.values())})

# Sort the DataFrame by count in descending order
production_df = production_df.sort_values('Count', ascending=False)
production_df
```

Task 10 Sort the data in descending order basedon revenue and filter the top 500 movies.Find the measures of central tendency for the following columnsusing the filtered data:

```python
sort_500=df.sort_values('revenue',ascending=False).head(500)
sort_500
```

```python
final_data=sort_500[['budget','revenue','runtime']] 
final_data
```

```python
Mean=[]
Median=[]
Mode=[]

for col in final_data.head():
    Mean.append(final_data[col].mean())
    Median.append(final_data[col].median())
    Mode.append(final_data[col].mode()[0])
index=['Mean','Median','Mode']
cols=['budget','revenue','runtime']
```

```python
Central_tend=pd.DataFrame([Mean,Median,Mode],columns=cols,index=index)
Central_tend
```

```python
## Calculating dispersion
Mean=[]
STD=[]
CV=[]

for col in final_data.head():
    Mean.append(final_data[col].mean())
    STD.append(final_data[col].std())
    CV.append(final_data[col].std()*100/final_data[col].mean())
    
index=['Mean','Standard Deviation','Coefficient of Variance']
cols=['budget','revenue','runtime']
```

```python
Measures_Desper=pd.DataFrame([Mean,STD,CV],columns=cols,index=index)
Measures_Desper
```

```python
# Outliers
for data in final_data.head():
    lo=[]
    uo=[]
    ser=pd.Series(final_data[data])
    minimum=min(ser)
    maximum=max(ser)
    ranges=maximum-minimum
    q1=ser.quantile(0.25)
    q3=ser.quantile(0.75)
    iqr=q3-q1
    uf=q3+1.5*iqr
    lf=q1-1.5*iqr
    
    for i in final_data[data]:
        if i<lf:
            lo.append(i)
        elif i>uf:
            uo.append(i)
    if data=='budget':
        budget=['budget',minimum,maximum,ranges,q1,q3,iqr,lf,uf,(lo,uo)]
    elif data=='revenue':
        revenue=['revenue',minimum,maximum,ranges,q1,q3,iqr,lf,uf,(lo,uo)]
    elif data=='runtime':
        runtime=['runtime',minimum,maximum,ranges,q1,q3,iqr,lf,uf,(lo,uo)]
        
    
```

```python
outliers=pd.DataFrame([budget,revenue,runtime],columns=
                      ['data','minimum','maximum','ranges','q1','q3','iqr','lf','uf','Outliers'])
outliers
```
